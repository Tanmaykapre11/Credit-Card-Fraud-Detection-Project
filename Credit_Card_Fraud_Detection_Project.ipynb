```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
data['NormalizedAmount'] = scaler.fit_transform(data['Amount'].values.reshape(-1,1))
```

```python
data.head()
```

```python
data.drop(['Amount','Time'], inplace=True, axis=1)
```

```python
data.head()
```

```python
X = data.drop(['Class'], axis=1)
y = data['Class']
```

```python
X
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
print('Shape of X_train', X_train.shape)
print('Shape of y_train', X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Decision_Tree = DecisionTreeClassifier()
```

```python
Decision_Tree.fit(X_train,y_train)
```

```python
DT_predict = Decision_Tree.predict(X_test)
```

```python
from sklearn.metrics import *
```

```python
DT_Acc_Score = accuracy_score(y_test, DT_predict)*100
```

```python
DT_Acc_Score
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Random_Forest = RandomForestClassifier(n_estimators=100)
```

```python
Random_Forest.fit(X_train,y_train)
```

```python
RF_Predict = Random_Forest.predict(X_test)
```

```python
RF_Acc_Score = accuracy_score(y_test, RF_Predict)*100
```

```python
RF_Acc_Score
```

```python
print('Accuracy Score of Decision Tree: ',DT_Acc_Score)
print('Accuracy Score of Random Forest: ',RF_Acc_Score)
```

```python
confusion_matrix_dt = confusion_matrix(y_test, DT_predict)
```

```python
def metrics(actuals, predictions):
    print("Accuracy: {:.5f}".format(accuracy_score(actuals, predictions)))
    print("Precision: {:.5f}".format(precision_score(actuals, predictions)))
    print("Recall: {:.5f}".format(recall_score(actuals, predictions)))
    print("F1-score: {:.5f}".format(f1_score(actuals, predictions)))
```

```python
metrics(y_test, DT_predict)
```

```python
metrics(y_test, RF_Predict)
```

# Here we can clearly see that Random Forest Model is giving best Results instead of Decision Tree.
